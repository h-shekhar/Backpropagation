```python
names = ['x','y','z','class']
data = pd.read_csv('train.csv', names=names)
print(data.shape)
print(data.head())
```

```python
# Standardize the data(mean=0, std-dev=1)
from sklearn.preprocessing import StandardScaler 
import numpy
array = data.values
# separate array into input and output components
features = array[:,0:3]
labels = array[:,3]
```

```python
labels = labels.reshape(10,1) 
labels.shape
```

```python
np.random.seed(42)  
weights = np.random.rand(3,1)  
bias = np.random.rand(1)  
lr = 0.05  
```

```python
def sigmoid(x):  
    return 1/(1+np.exp(-x))
```

```python
def sigmoid_der(x):  
    return sigmoid(x)*(1-sigmoid(x))
```

```python
for epoch in range(20000):  
    inputs = features

    # feedforward step1
    XW = np.dot(features, weights) + bias

    #XW = nparray(np.dot(features, weights),dtype=np.float32) + bias
    #z=np.array(np.dot(X, T),dtype=np.float32)

    #feedforward step2
    z = sigmoid(XW)


    # backpropagation step 1
    error = z - labels

    print(error.sum())

    # backpropagation step 2
    dcost_dpred = error
    dpred_dz = sigmoid_der(z)

    z_delta = dcost_dpred * dpred_dz

    inputs = features.T
    weights -= lr * np.dot(inputs, z_delta)

    for num in z_delta:
        bias -= lr * num
```

```python
single_point = np.array([1,0,0])  
result = sigmoid(np.dot(single_point, weights) + bias)  
print(result) 
```
